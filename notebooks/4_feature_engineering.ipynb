```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/Walmart.csv", parse_dates=['Date'])
```

```python
df = df.sort_values(['Store', 'Date']).reset_index(drop=True)
```

```python
df.columns
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week.astype(int)
```

```python
df['Holiday'] = df['Holiday_Flag'].map({0: 'No', 1: 'Yes'})
```

```python
df = df.sort_values(['Store', 'Date']).reset_index(drop=True)
```

```python
df['Lag_1_Week'] = df.groupby('Store')['Weekly_Sales'].shift(1)
df['Lag_2_Week'] = df.groupby('Store')['Weekly_Sales'].shift(2)
df['Lag_4_Week'] = df.groupby('Store')['Weekly_Sales'].shift(4)
```

```python
df = df.dropna().reset_index(drop=True)
```

Feature Engineering Summary
Time-based features (year, month, week) and lagged sales variables were engineered to capture seasonality and historical sales patterns. These features prepare the dataset for predictive modeling and time-series forecasting.
